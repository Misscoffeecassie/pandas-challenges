```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

```python
 # Calculate the total number of schools
total_schools=school_data_complete["school_name"].nunique()
total_schools
```

```python
# Calculate the total number of students
total_students=student_data["Student ID"].count()
total_students
```

```python
# Calculate the total budget
total_budget=school_data["budget"].sum()
total_budget
```

```python
# Calculate the average maths score 
avg_maths=student_data["maths_score"].mean()
avg_maths
```

```python
# Calculate the average reading score
avg_reading=student_data["reading_score"].mean()
avg_reading
```

```python
# Calculate the percentage of students with a passing maths score (50 or greater)

percentage_maths_pass=student_data[(student_data["maths_score"]>=50)].shape[0]/ total_students*100
percentage_maths_pass
```

```python
# Calculate the percentage of students with a passing reading score (50 or greater)
percentage_reading_pass=student_data[(student_data["reading_score"]>=50)].shape[0]/ total_students*100
percentage_reading_pass
```

```python
#Calculate the percentage of students who passed maths **and** reading (% Overall Passing)
percentage_math_reading_pass=student_data[(student_data["maths_score"]>=50) & (student_data["reading_score"]>=50)].shape[0]/ total_students*100
percentage_math_reading_pass
```

```python
# Create a dataframe to hold the above results
# Optional: give the displayed data cleaner formatting
area_summary={"Total Schools":[total_schools],
              "Total Students":[total_students], 
              "Total Budget($)":[total_budget], 
              "Average Maths Score": [round(avg_maths, 2)],
              "Average Reading Score":[round(avg_reading,2)],
              "% Passing Maths":[round(percentage_maths_pass,2)],
              "% Passing Reading":[round(percentage_reading_pass,2)],
              "% Overall Passing": [round(percentage_math_reading_pass,2)] }
area_summary=pd.DataFrame(area_summary)

area_summary
```

```python
school_data
```

```python
# Display School Name & School Type & total budget & total students per school
school_new_data=pd.DataFrame()
school_new_data[["School Name", "School Type", "Total Students", "Total School Budget"]]=school_data[["school_name", "type", "size", "budget"]]
school_new_data["Per Student Budget"]=school_new_data['Total School Budget']/school_new_data['Total Students']

#school_new_data["Average Maths Score"]=school_data_complete.loc[(school_data_complete['school_name']==school_new_data['School Name'])]
school_new_data=school_new_data.sort_values('School Name', ascending=True)
school_new_data
```

```python
#  Average Maths & reading Score Per School
avg_maths_reading_score=school_data_complete.groupby('school_name')['maths_score','reading_score'].mean().reset_index()
avg_maths_reading_score=avg_maths_reading_score.rename(columns={'school_name':"School Name", 'maths_score':"Average Maths Score", 'reading_score':"Average Reading Score"})
avg_maths_reading_score
```

```python
school_new_data
```

```python
per_school_summary=school_new_data.merge(avg_maths_reading_score,on=['School Name'], how='outer')
per_school_summary
```

```python
# % Passing Maths
grouped_passing_maths=school_data_complete.loc[school_data_complete['maths_score']>=50].groupby('school_name')
# grouped_passing_maths
passing_maths_count=grouped_passing_maths['school_name'].count()
# passing_maths_count
passing_maths_percentage=(passing_maths_count.values/ per_school_summary['Total Students']*100).to_frame()
passing_maths_percentage
per_school_summary['% Passing Maths']=passing_maths_percentage
per_school_summary
```

```python
  # % Passing Reading
grouped_passing_reading=school_data_complete.loc[school_data_complete['reading_score']>=50].groupby('school_name')
# grouped_passing_reading
passing_reading_count=grouped_passing_reading['school_name'].count()
# passing_reading_count
passing_reading_percentage=(passing_reading_count.values/ per_school_summary['Total Students']*100).to_frame()
passing_reading_percentage
per_school_summary['% Passing Reading']=passing_reading_percentage
per_school_summary
```

```python
  # % Overall Passing (The percentage of students that passed maths **and** reading.)
grouped_passing_overall=school_data_complete.loc[(school_data_complete['reading_score']>=50) & (school_data_complete['maths_score']>=50)].groupby('school_name')
# grouped_overall_reading
passing_overall_count=grouped_passing_overall['school_name'].count()
# passing_overall_count
passing_overall_percentage=(passing_overall_count.values/ per_school_summary['Total Students']*100).to_frame()
passing_overall_percentage
per_school_summary['% Overall Passing']=passing_overall_percentage
per_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python

spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]


data_cut=pd.cut(per_school_summary['Per Student Budget'],bins=spending_bins,labels=labels)
school_spending_avgmaths=per_school_summary.groupby(data_cut)['Average Maths Score'].mean()
school_spending_avgreading=per_school_summary.groupby(data_cut)['Average Reading Score'].mean() 
school_spending_permaths=per_school_summary.groupby(data_cut)['% Passing Maths'].mean()
school_spending_perreading=per_school_summary.groupby(data_cut)['% Passing Reading'].mean()
school_spending_peroverall=per_school_summary.groupby(data_cut)['% Overall Passing'].mean()
spending_summary=pd.DataFrame({
                                 'Average Maths Score':school_spending_avgmaths,
                                 'Average Reading Score':school_spending_avgreading,
                                '% Passing Maths':school_spending_permaths,
                                '% Passing Reading':school_spending_perreading,
                                '% Overall Passing': school_spending_peroverall})
spending_summary
```
